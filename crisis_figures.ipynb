# Dependence of the recovery time after critical jumps in the stable system on the jump size

```python
tcrit = 500 # timestep of jump

jumpsize = 2000 # get from data
threshold = 10 #10 #jumpsize / 100 # 0.01 # jumpsize / N
#threshold = 0.05 * jumpsize #5% of jumpsize
max_time = 2000



q_crit = np.mean(q_evol[tcrit-50:tcrit-1, 0])
print(q_crit)

t = tcrit+1
while (t < max_time and q_evol[t, 0] > q_crit + threshold):
    t += 1

print(t)
print(q_evol[t,0])
print(q_evol[t+1,0])
print(q_evol.shape)

def recovery_time(q_evol, tcrit, jumpsize, idx=0, threshold=10):
    max_time = q_evol.shape[0]
    q_crit = np.mean(q_evol[tcrit-50:tcrit-1, idx])
    t = tcrit+5

    if idx != 0:
        while (t < max_time and (q_evol[t, idx] < q_evol[t+1, idx])):
            t += 1

    while (t < max_time and (q_evol[t, idx] > q_crit + threshold)):
        t += 1

    return t-tcrit

def recovery_time_scaled(q_evol, tcrit, jumpsize, idx=0, tol = 0.01):
    max_time = q_evol.shape[0]
    q_crit = np.mean(q_evol[tcrit-50:tcrit-1, idx])
    t = tcrit+5
    threshold = jumpsize * tol

    if idx != 0:
        while (t < max_time and (q_evol[t, idx] < q_evol[t+1, idx])):
            t += 1

    while (t < max_time and (q_evol[t, idx] > q_crit + threshold)):
        t += 1

    return t-tcrit
```

### description how obtained the figure
Setting: 
- N = 100, G = 5
- c1 + .. + c5 = 0.3, ci random, m = 0.5, s = 0.1
- 10 trials for different jumpsizes = 100, 200, ..., 2000 (first glob param. increases by this size at t = 500)

Curve:

time until the global parameters are back at the previous equilibria. 

Computation of recovery time:

For each global parameter: t_recov = "t where g(t) first becomes smaller than g(tcrit)" - tcrit

Curve: mean over all global params of recovery time, error band = percentiles
